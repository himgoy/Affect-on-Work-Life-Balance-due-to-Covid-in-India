```python
import pandas as pd
```

```python
scweet = pd.read_csv('/content/drive/MyDrive/1 Work Life Balance/All Tweets.csv')
scweet.head()
```

```python
tweet_id=scweet['Tweet URL'].str.extract(r'(\d{19})')
tweet_id=tweet_id[0].to_list()
for i in range(0, len(tweet_id)):
    tweet_id[i] = int(tweet_id[i])
print('Length of tweet_id ' + str(len(tweet_id)))
tweet_id[:4]
```

```python
IEEE_data = pd.read_csv('/content/drive/MyDrive/1 Work Life Balance/WLB_Full_Tweet_Data.csv')
IEEE_data.head()
```

```python
IEEE_Tweet_Ids = IEEE_data.id.to_list()
print('Length of IEEE_tweet_id ' + str(len(IEEE_Tweet_Ids)))
IEEE_Tweet_Ids[:4]
```

```python
All_Tweet_IDs= tweet_id+IEEE_Tweet_Ids
print('Length of All_tweet_id ' + str(len(All_Tweet_IDs)))
All_Tweet_IDs[:4]
```

```python
print('Number of Tweet IDs : '+str(len(All_Tweet_IDs)))
All_Tweet_IDs = set(All_Tweet_IDs)  # to find unique ids
All_Tweet_IDs = list(All_Tweet_IDs)
print('Number of Unique Tweet IDs : '+str(len(All_Tweet_IDs)))

# print('Number of Tweet IDs : '+str(len(tweet_id)))
# tweet_id = set(tweet_id)  # to find unique ids
# tweet_id = list(tweet_id)
# print('Number of Unique Tweet IDs : '+str(len(tweet_id)))
```

```python
with open('/content/drive/MyDrive/1 Work Life Balance/All_Tweet_IDs.txt', 'w') as f:
    for item in All_Tweet_IDs:
        f.write("%s\n" % item)
```
